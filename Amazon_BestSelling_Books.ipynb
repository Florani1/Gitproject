```python
genre = bs.find_all('div',{'class':"kc-sublist"}) #kc-nonfiction kc-sublist-part-right kc-sublist-part-active
genre
```

```python
genre_names = []

for element in genre:
    # Find the first 'div' element within the current element
    div_element = element.find('div')

    # Check if a 'div' element was found
    if div_element:
        # Append the text content of the 'div' element to the 'rank' list
        genre_names.append(div_element.text.strip())
    else:
        # Handle the case where no 'div' element was found
        genre_names.append("N/A")
genre_names
```
